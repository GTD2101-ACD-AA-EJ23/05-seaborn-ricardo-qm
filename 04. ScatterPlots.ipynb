# Gráficos de Dispersión

En este JNB se revisará el tema de Gráficos de Dispersión.

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
insurance_filepath = "https://raw.githubusercontent.com/GTD2101-ACD-AA-EJ23/05-seaborn-ricardo-qm/main/data/insurance.csv"

# Read the file into a variable insurance_data
insurance_data = pd.read_csv(insurance_filepath)
```

```python
insurance_data.head()
```

# Gráficos de Dispersión

Se puede mostrar gráficamente la relación que hay entre dos variables con un gráfico de dispersión.

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```
